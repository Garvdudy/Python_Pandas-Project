# Data Cleaning in Panda

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\PRERNA\OneDrive - George Brown College\Desktop\Data Anylist\Pandas\Material\Customer Call List.xlsx")
df
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df["Last_Name"].str.lstrip("...")
df["Last_Name"].str.lstrip("/")
df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df

# NOTE : You can use one bye one if there is one type of error or use the last statement to do in one-line-code.....
```

```python
# This [^a-zA-Z0-9] is a format to remove anything expect a-z, A-Z and 0-9.....
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex = True)

# This line of code will make sure to change int to str to help in adding sepraters
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

# This is how we add seprators
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

# This two line of code is to replace strings
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')

df
```

```python
df[["Street_address", "State", "Zip_code"]] = df["Address"].str.split(',', expand = True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('Y', 'Yes')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df["Paying Customer"] = df["Paying Customer"].str.replace('N', 'No')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Y', 'Yes')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('N', 'No')
df
```

```python
df = df.replace('N/a', '')
df = df.replace('NaN', '')
df = df.replace('No/a', '')
df = df.replace('--', '')

# This fillna is use to fil the NaN values or Any blank Values...
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Yes':
        df.drop(x, inplace = True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)

df

# Another way to drop null values...
# df = df.dropna((subset = "Phone_Number"), inplace = True)
```
